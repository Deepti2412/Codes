```python
emoticons_happy = set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'
    ])
emoticons_sad = set([
    ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('
    ])
emoticons=emoticons_happy.union(emoticons_sad)
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)        
    return input_txt

def clean_tweets(lst):
    # remove twitter Return handles (RT @xxx:)
    lst = np.vectorize(remove_pattern)(lst, "RT @[\w]*:")
    # remove twitter handles (@xxx)
    lst = np.vectorize(remove_pattern)(lst, "@[\w]*")
    # remove URL links (httpxxx)
    lst = np.vectorize(remove_pattern)(lst, "https?://[A-Za-z0-9./]*")   
     # Remove all single characters 
    lst = np.vectorize(remove_pattern)(lst, "\s+[a-zA-Z]\s+") #space followed by alphabet, followed by space
    # remove numbers
    lst = np.vectorize(remove_pattern)(lst, "\d")
    # remove punctuations
    lst = np.vectorize(remove_pattern)(lst, "[\!\:\&\;\#\@//]")
    return lst
    lst = np.vectorize(remove_pattern)(lst, "[0-9]{2}/[0-9]{2}/[0-9]{4}")
    return lst
    lst = np.vectorize(remove_pattern)(lst, "[0-9]{2}:[0-9]{2}]")
    return lst
    lst = np.vectorize(remove_pattern)(lst, "\\r\\n|\\r|\\n")
    return lst
    lst = np.vectorize(remove_pattern)(lst, "[\-/ /+\]")
    return lst
```

```python
patterns = r"""
  NP: {<DT|JJ|NN.*>+}          # Chunk sequences of DT, JJ, NN
  PP: {<IN><NP>}               # Chunk prepositions followed by NP
  VP: {<VB.*><NP|PP|CLAUSE>+$} # Chunk verbs and their arguments
  CLAUSE: {<NP><VP>}           # Chunk NP, VP
  """
```

```python
chunkparser = nltk.RegexpParser(patterns)
tree = chunkparser.parse(nltk.pos_tag(filtered_tokens))
print(tree)
```
